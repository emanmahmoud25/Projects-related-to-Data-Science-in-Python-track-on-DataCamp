```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
data=pd.read_csv('nobel.csv')
data
```

```python
#most commonly awarded gender and birth country
top_gender=data['sex'].mode()[0]
top_country=data['birth_country'].mode()[0]
print(top_country)
print(top_gender)
```

```python
#highest ratio of US-born Nobel Prize winners to total winners in all categories
data['Decade'] = (data['year'] // 10) * 10
'''This line of code performs integer division of the "year" values by 10. Integer division (//) rounds down the result to the nearest integer. This effectively rounds down the year to the nearest decade. For example:
2023 // 10 = 202
1994 // 10 = 199
2019 // 10 = 201
Multiply by 10 to get the decade value:
python
Copy
(data['year'] // 10) * 10
This line of code takes the result of the integer division from the previous step and multiplies it by 10 to get the actual decade value. For example:
202 * 10 = 2020
199 * 10 = 1990
201 * 10 = 2010'''
data_united=data[data['birth_country']=='United States of America']
us_born_ratio=data_united.groupby('Decade').size()/data.groupby('Decade').size()
max_decade_usa=us_born_ratio.idxmax()
max_decade_usa
```

```python
#decade and Nobel Prize category combination had the highest proportion of female
female_Data=data[data['sex']=='Female']
female_proportion=female_Data.groupby(['Decade','category']).size()/data.groupby(['Decade','category']).size()
max_female=female_proportion.idxmax()
max_female_dict={max_female[0]:max_female[1]}
max_female_dict
```

```python
#first woman to receive a Nobel Prize, and in what category
first_woman=data[data['sex']=='Female'].sort_values(by='year').iloc[0]
#asending is default order ,i want first woman that mean i want from From oldest to newest
first_woman_name=first_woman['full_name']
first_woman_category=first_woman['category']
print(first_woman_name)
print(first_woman_category)
```

```python
#individuals or organizations have won more than one Nobel Prize throughout the years
repeat_winners=data.groupby('full_name' ).filter(lambda x: len(x)>1)
repeat_list=repeat_winners['full_name' ].unique().tolist()
repeat_list
```
