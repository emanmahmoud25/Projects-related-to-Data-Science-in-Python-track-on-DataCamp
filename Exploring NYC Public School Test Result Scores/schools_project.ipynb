```python
import pandas as pd
data=pd.read_csv('schools.csv')
data
```

```python
best_math_schools=data[data['average_math'] >=0.8*800]
best_math_schools=best_math_schools[['school_name','average_math']]
best_math_schools=best_math_schools.sort_values(by='average_math' ,ascending=False)
print(best_math_schools)
```

```python
data['total_SAT']=data[['average_math','average_reading','average_writing']].sum(axis=1)
top_10_schools=data[['school_name','total_SAT']].sort_values(by='total_SAT' ,ascending=False).head(10)
top_10_schools
```

```python
borough_stats=data.groupby('borough').agg(num_schools=('school_name','count'),
                                         average_SAT=('total_SAT','mean'),
                                         std_SAT=('total_SAT','std')).reset_index()
borough_stats
```

```python
largest_std_dev=borough_stats[borough_stats['std_SAT']== borough_stats['std_SAT'].max()]
largest_std_dev
```

```python
largest_std_dev=largest_std_dev.round({'average_SAT':2 ,'std_SAT':2})
largest_std_dev
```
